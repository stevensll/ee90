```python
import csv
import matplotlib.pyplot as plt

def graph_plot(file_name):
    times = []
    temps = []
    with open(file_name, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            time = float(row['Time'])
            temp = float(row['Temperature'])
            times.append(time)
            temps.append(temp)

    plt.figure(figsize=(10, 5))
    plt.plot(times, temps, marker='o', linestyle='None', markersize=6)
    plt.xlabel('Time (s)')
    plt.ylabel('Temperature (C)')
    plt.grid(True)
    plt.show()
```

```python
graph_plot("on_off_good.csv")
graph_plot("pid_good_2hr.csv")
graph_plot("pid_good_30min.csv")
```
