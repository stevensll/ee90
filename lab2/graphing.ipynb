```python
import csv
import matplotlib.pyplot as plt

def graph_plot(file_name):
    times = []
    temps = []
    with open(file_name, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            time = float(row['Time'])
            temp = float(row['Temperature'])
            times.append(time)
            temps.append(temp)

    plt.figure(figsize=(10, 5))
    plt.plot(times, temps, marker='o', linestyle='None', markersize=6)
    plt.xlabel('Time (s)')
    plt.ylabel('Temperature (K)')
    plt.grid(True)
    plt.show()
```

```python
def graph_plot_region(file_name, low_time, max_time, setpoint):
    times = []
    temps = []
    with open(file_name, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            time = float(row['Time'])
            temp = float(row['Temperature'])
            if time > low_time and time < max_time:
                times.append(time)
                temps.append(temp)
    plt.figure(figsize=(10, 5))
    plt.axhline(y=setpoint, color='r', linestyle='--', label=f'Setpoint: {setpoint} K')
    plt.axhline(y=setpoint + 0.1, color='g', linestyle='--', label=f'+0.1 K bound: {setpoint + 0.1:.3f} K')
    plt.axhline(y=setpoint - 0.1, color='g', linestyle='--', label=f'-0.1 K bound: {setpoint - 0.1:.3f} K')

    plt.plot(times, temps, marker='o', linestyle='None', markersize=6)
    plt.xlabel('Time (s)')
    plt.ylabel('Temperature (K)')
    plt.legend()  # ðŸ‘ˆ This shows the "key"/legend

    plt.grid(True)
    plt.show()
```

```python
graph_plot_region("pid_good_30min.csv", 60 * 9, 60*11, 300)
graph_plot_region("pid_good_30min.csv", 60 * 29, 60*30, 300)
graph_plot_region("pid_good_2hr.csv", 60 * 55, 60*60, 310.161)
graph_plot_region("pid_good_2hr.csv", 60 * 115, 60*120, 303.388)
```
